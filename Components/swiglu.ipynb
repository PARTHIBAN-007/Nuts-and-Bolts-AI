```python
import numpy as np

x = np.array([[0.8,-0.3,0.5,1.2]])
w_a = np.array([[0.5,-0.2],
                [0.8,0.1],
                [-0.7,0.4],
                [0.2,0.9]
                ])
w_b = np.array([
    [2.5,-3.0],
    [-0.1,0.2],
    [0.3,-0.5],
    [1.0,0.8]
])
print(x.shape)
print(w_a.shape)
print(w_b.shape)
```

```python
a = np.dot(x,w_a)
a
```

```python
b = np.dot(x,w_b)
b
```

```python
def sigmoid(x):
    return 1 / (1+ np.exp(-x))
sigmoid_b = sigmoid(b)
sigmoid_b
```

```python
swish = []
for b_val,sig_val in zip(b,sigmoid_b):
    swish.append(b_val*sig_val)
print(swish)
```

```python
output = []
for a_val,swish_val in zip(a,swish):
    output.append(a_val*swish_val)
print(output)
```
