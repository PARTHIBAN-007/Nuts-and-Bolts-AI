```python
import numpy as np

x = np.array([0.8,-0.3,0.5,1.2])
w_a = np.array([[0.5,-0.2],[0.8,0.1],[-0.7,0.4],[0.2,0.9]])
w_b = np.array([[0.9,-0.4],[-0.1,0.6],[0.3,-0.8],[0.7,0.2]])

print(x.shape)
print(w_a.shape)
print(w_b.shape)
```

```python
a = np.dot(x,w_a)
a
```

```python
b = np.dot(x,w_b)
b
```

```python
def sigmoid(x):
    return 1 / (1+ np.exp(-x))

gate = sigmoid(b)
gate
```

```python
output = []
for a_val,gate_val in zip(a,gate):
    output.append(a_val*gate_val)
output
```
