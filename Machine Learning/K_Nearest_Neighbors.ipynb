```python
import numpy as np


x = [
    [1.2,2,0],
    [1.8,3,0],
    [2.1,2,0],
    [2.5,4,1],
    [3.2,3,1],
    [3.8,5,1]
]
new_h = [2.3,3]
k_value = 3
```

```python
d = []
n = len(x)
for i in range(n):
    dist = (new_h[0] - x[i][0] )**2 + (new_h[1]-x[i][1])**2
    d.append((dist**0.5,i))


for dist in d:
    print(dist)
```

```python
d.sort()
d
```

```python
top_k = []
for i in range(k_value):
    idx = d[i][1]
    top = x[idx].append(d[i][0])
    top_k.append(x[idx])

top_k
```

```python
from collections import defaultdict
nums = []

for i in range(k_value):
    nums.append(top_k[i][-1])
nums
```

```python
from collections import Counter

def mode(values):
    counts = Counter(values)
    max_freq = max(counts.values())
    modes = [k for k, v in counts.items() if v == max_freq]
    return modes

mode(nums)
```
