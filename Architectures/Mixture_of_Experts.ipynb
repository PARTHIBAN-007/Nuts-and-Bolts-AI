```python
import torch
import torch.nn.functional as F
```

```python
x = torch.tensor([
    [0.8,-0.2,0.5,1.2]
])

router_weights = torch.tensor([
    [1.2,-0.5,0.8,0.3,-1.5,0.6,0.9,-0.1],
    [-0.9,1.8,-0.2,1.10,0.4,-1.2,0.3,0.7],
    [0.4,0.7,1.5,-0.8,0.9,0.1,-1.10,0.5],
    [0.6,-1.10,0.4,1.30,-0.7,1.4,0.2,-0.9]
])

print(x.shape)
print(router_weights.shape)
```

```python
logits = x @ router_weights
logits
```

```python
expert_1 = torch.tensor([[0.1,0.2,0.3,0.4]])
expert_6 = torch.tensor([[2.1,2.2,2.3,2.4]])
print(expert_1.shape)
print(expert_6.shape)
```

```python
topk_vals, topk_idx = torch.topk(logits,2,dim=-1)
print(topk_vals,topk_idx)
```

```python
gate_weights = F.softmax(topk_vals,dim=-1)
gate_weights
```

```python
output = expert_1*gate_weights[0][1]+expert_6*gate_weights[0][0]
output
```
